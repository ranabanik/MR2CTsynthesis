```python
import os
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
dir_log='C:\\Users\\Reasat\\Projects\\MR2CTsynthesis\\log'
```

```python
filepath_log=os.path.join(dir_log,'2019-05-10-23-37-05.bin')
log=pickle.load(open(filepath_log,'rb'))
print(log.keys())
```

```python
plt.figure(figsize = [20,10])
for key in log.keys():
    plt.plot(log[key],label=key)
plt.legend()
```

```python
plt.figure(figsize = [20,10])
for key in log.keys():
    if key in ['loss_d_fake_epoch_train','loss_d_real_epoch_train','loss_g_d_epoch_train']:
        plt.plot(log[key],label=key)
plt.legend()
```
